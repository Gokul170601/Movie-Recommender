```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/Gokul170601/Movie-Recommender/main/IMDB_data.csv')
data.head()
```

```python
data.info()
```

```python
df=data[[ 'movie_imdb_link','movie_title','director_name','actor_1_name', 'actor_2_name',
       'actor_3_name', 'genres',  'plot_keywords', ]]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['movie_id'] = df['movie_imdb_link'].apply(lambda x: x.split('/')[4],)
```

```python
df['actors']=df['actor_1_name']+','+df['actor_2_name']+','+df['actor_3_name']
```

```python
df.drop(columns=['movie_imdb_link','actor_1_name','actor_2_name','actor_3_name'],inplace=True)
```

```python
df['genres']=df['genres'].str.replace('|',' ').str.lower()
df['plot_keywords']=df['plot_keywords'].str.replace('|',' ').str.lower()
df['actors']=df['actors'].str.replace(' ','').str.replace(',',' ').str.lower()
df['director_name']=df['director_name'].str.replace(' ','').str.lower()
```

```python
df.head()
```

```python
df['tags']=df['genres']+' '+df['plot_keywords']+' '+df['actors']+' '+df['director_name']
df.drop(columns=['genres','plot_keywords','actors','director_name'],inplace=True)
```

```python
df.head(5)
```

```python
df['movie_title']=df['movie_title'].str.strip()
```

```python
df.to_csv('processed.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf=TfidfVectorizer()
model=tfidf.fit_transform(df['tags'])
```

```python
model.shape
```

```python
def recommend(movie):
  idx=df[df['movie_title']==movie].index[0]
  similarity=cosine_similarity(model[idx],model)
  sorted_list=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
  for i in sorted_list[1:6]:
    print(df.iloc[i[0]].movie_title)
```

```python
recommend('Spider-Man 3')
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
```
